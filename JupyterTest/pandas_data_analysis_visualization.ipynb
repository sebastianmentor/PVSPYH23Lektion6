# Dataanalys och visualisering i Pandas
I den här notebooken kommer vi att utforska några grundläggande dataanalys och visualiseringstekniker med Pandas och Matplotlib.

## Dataanalys

```python
import pandas as pd

# Skapar en exempel DataFrame
data = {
    'Produkt': ['Äpple', 'Banan', 'Apelsin', 'Päron', 'Kivi'],
    'Pris': [10, 20, 15, 25, 33], # [10, 15, 20, 25, 33] 
    'Sålda': [34, 54, 23, 46, 20]
}
df = pd.DataFrame(data)

# Grundläggande statistik
df.describe()
```

## Visualisering med Matplotlib

```python
import matplotlib.pyplot as plt

# Barplot för antalet sålda produkter
plt.figure(figsize=(8, 4))
plt.bar(df['Produkt'], df['Sålda'], color='red')
plt.title('Antal Sålda Produkter')
plt.xlabel('Produkt')
plt.ylabel('Sålda')
plt.show()
```

```python
import numpy as np

plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
