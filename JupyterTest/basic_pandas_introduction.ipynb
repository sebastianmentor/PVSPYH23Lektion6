## Pandas Series

```python
import pandas as pd

# Skapar en Series
data = [1, 3, 5, 7, 9]
ser = pd.Series(data)
ser
```

## Indexering och urval i Series

```python
# Hämta elementet på index 2
element = ser[2]
element
```

## Pandas DataFrame

```python
# Skapar en DataFrame
data = {
    'År': [2019, 2020, 2021, 2022],
    'Intäkter': [10, 20, 30, 40]
}
df = pd.DataFrame(data)
df
```

```python
new_data = [[1,2,3],[4,5,6],[7,8,9]]

df_2 = pd.DataFrame(new_data, columns=['ålder', 'läng','betyg'])
df_2
```

## Indexering och urval i DataFrame

```python
# Efter kolumnnamn
# df['År']

# Efter radindex
# df.iloc[0]


# # Slice
df[0:2]
```

## Grundläggande operationer med DataFrame

```python
# Lägger till en ny kolumn
df['Nytt värde'] = [100, 200, 300, 222]
# df['Nytt värde'].dtype 
# df['År'].dtype

# Filtrera ut rader
filtered_df = df[df['Intäkter'] > 20]
filtered_df
```
