## Hantera saknade värden

```python
# Lägger till några saknade värden
df.loc[2, 'Stad'] = None
df
# Hantera saknade värden
df_filled = df.fillna('Okänd')  # Ersätter saknade värden med 'Okänd'
df_dropped = df.dropna(ignore_index=True)        # Tar bort rader med saknade värden
df_filled, df_dropped
```

## Datarening och förändring

```python
# Skapar en ny kolumn
df['Ålder + 10'] = df['Ålder'] + 10

# Ändrar en befintlig kolumn
df['Stad'] = df['Stad'].str.upper()  # Konverterar alla stadsnamn till versaler
df
```
